## Work Style and Interests

We could see the user's interests and their work style preference through a survey type input and based on those reccomend careers. For example, how important is achieving something? How much of a leader are they, their stress tolerance, etc...

```python
styles.shape
```

```python
styles['Title'].nunique()
```

```python
styles['Attribute Type'] = 'Work Styles'
styles
```

```python
columnReorder = ['O*NET-SOC Code', 'Title', 'Attribute Type', 'Element ID', 'Element Name', 'Scale Name', 'Data Value']
styles = styles[columnReorder]
styles.head()
```

```python
values = pd.read_excel('Work Values.xlsx')
values.head()
```

```python
values.shape
```

```python
values['Title'].nunique()
```

```python
values['Attribute Type'] = 'Work Values'
values.head()
```

```python
columnReorder = ['O*NET-SOC Code', 'Title', 'Attribute Type', 'Element ID', 'Element Name', 'Scale Name', 'Data Value']
values = values[columnReorder]
values.head()
```

```python
work = pd.concat([values, styles], ignore_index=True)
work.head()
```
