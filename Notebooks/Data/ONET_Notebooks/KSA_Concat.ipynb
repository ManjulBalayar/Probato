## Converting into a Long Format

The goal is to combine these different dataframes: knowledge_df, skills_df, abilities_df, into a singular df for convenience for the model. Is this a good idea since a model might be only able to take 1 model. Not to mention I have 10+ more datasets and will probably join them all together. Problem is there are different numbers of knowledge, skills, abilities and each have their own unique data value for `Level` and `Importance` so a wide format wouldnt really work, so I was thinking maybe a long format.

Example:

| O*NET-SOC Code | Title           | Attribute Type | Element ID | Element Name      | Scale Name  | Data Value |
|----------------|-----------------|----------------|------------|-------------------|-------------|------------|
| 432            | Chief Executive | Knowledge      | 1          | Telecommunications | Level       | 0.93       |
| 432            | Chief Executive | Knowledge      | 1          | Telecommunications | Importance  | 1.56       |
| 432            | Chief Executive | Skill          | 2.A.1      | Active Listening   | Level       | 4.5        |
| 432            | Chief Executive | Skill          | 2.A.1      | Active Listening   | Importance  | 4.9        |
| 432            | Chief Executive | Ability        | 1.A.1      | Oral Comprehension | Level       | 3.8        |
| 432            | Chief Executive | Ability        | 1.A.1      | Oral Comprehension | Importance  | 4.7        |


```python
# Adding a 'Attribute Type' column
knowledge_df['Attribute Type'] = 'Knowledge'
skills_df['Attribute Type'] = 'Skill'
abilities_df['Attribute Type'] = 'Ability'
```

```python
# Concatenate the dataframes
ksa_df = pd.concat([knowledge_df, skills_df, abilities_df], ignore_index=True)
ksa_df.head()
```

```python
columnReorder = ['O*NET-SOC Code', 'Title', 'Attribute Type', 'Element ID', 'Element Name', 'Scale Name', 'Data Value']
ksa_df = ksa_df[columnReorder]
ksa_df.tail()
```
